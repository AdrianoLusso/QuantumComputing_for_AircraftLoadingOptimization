# Creating QUBO of Aircraft Loading Optimization (ALO) with slack variables

### Imports and utils

```python
import docplex
from docplex.mp.model import Model

from openqaoa.problems.converters import FromDocplex2IsingModel # function to translate docplex problems into Ising Hamiltonians
from openqaoa import QAOA, QUBO
from openqaoa.backends import create_device
from openqaoa.utilities import bitstring_energy

import numpy as np
```

```python
def visualize_results(results,num_c,num_p):
    for r in res:
        aux = ''
        for index,car in enumerate(r[:num_c*num_p]):
            if index % num_p == 0:
                aux += ' '+car
            else:
                aux += car
        print(aux)
```

### Description of container sizes and its features

T1 - medium --- $t_i=1$ --- $d_i=1$

T2 - small --- $t_i=1$ --- $d_i=0.5$

T2 - big --- $t_i=0.5$ --- $d_i=1$

* Para T1, toda la proporcion de su peso esta en una posicion ($t_i$), y tambien ocupa una posicion entera ($d_i$).
* Para T2, toda la proporcion de su peso esta en una posicion ($t_i$),pero ocupa la mitad de una posicion porque es chiquitito ($d_i$).
* Para T3, la mitad de la proporcion de su peso esta en una posicion y la otra mitad en otra ($t_i$),y ocupa una posicion entera ($d_i$).

## Creates constrained optimization formulation

```python
mdl = Model("ALO")
x = mdl.binary_var_list(len(allBinaryVariables), name="x")

# OBJECTIVE FUNCTION + CONTIGUITY OF BIG CONTAINERS
mdl.minimize(
    # OBJECTIVE FUNCTION
    -mdl.sum(
        t_i * m_i * x[len(positions) * i + j] 
        for j in positions
        for t_i,m_i,i in zip(t_list,containers_weight,containers)
    )
    +40* # CONTIGUITY OF BIG CONTAINERS --- P_C=40 is the penalty value for this penalty.
    mdl.sum(
        0.5 * mdl.sum(
                x[len(positions) * i + j]
                for j in positions
            )
        -
        mdl.sum(
            x[len(positions) * i + j] * x[len(positions) * i + (j+1)]
            for j in positions[:-1]
        )
        for i in containers
        if containers_type[i]==3
    )
)

# NO OVERLAPS --- its penalty value will be P_O
for j in positions:
    mdl.add_constraint(mdl.sum(
        d_i * x[len(positions) * i + j]
        for d_i,i in zip(d_list,containers) 
        ) <= 1
        ,ctname="no overlaps"
    )

# NO DUPLICATES --- its penalty value will be P_D. It will have to guarantee that P_D > 2*P_C
for i in containers:
    t_i = t_list[i]
    mdl.add_constraint(t_i * mdl.sum(
        x[len(positions) * i + j]
        for j in positions 
        ) <= 1
        ,ctname="no duplicates"
    )
    
'''
# CONTIGUITY OF BIG CONTAINERS
for i in containers:
    if containers_type[i]!=3:
        continue
    mdl.add_constraint(
        0.5 * mdl.sum(
                x[len(positions) * i + j]
                for j in positions
            )
        ==
        mdl.sum(
            x[len(positions) * i + j] * x[len(positions) * i + (j+1)]
            for j in positions[:-1]
        ) 
        ,ctname="contiguity"
    )
'''

# MAXIMUM CAPACITY --- its penalty value will be P_W
mdl.add_constraint(

    mdl.sum(
        t_i * m_i * x[len(positions) * i + j] 
        for j in positions
        for t_i,m_i,i in zip(t_list,containers_weight,containers)
    ) <= max_weight
    ,ctname="maximum capacity"
)
```

### Docplex model

```python
print(mdl.objective_expr,'\n')
print('constraints:')
for c in mdl.iter_constraints():
    print(c)
```

### Creates the Ising formulation

```python
mults=100 # this is the penalty value for P_O, P_D and P_W
converterWrapper = FromDocplex2IsingModel(mdl, unbalanced_const=False,multipliers=mults)

# gets the ising and qubo docplex
ising = converterWrapper.ising_model
mdl_ising = converterWrapper.qubo_docplex

# plot the ising
#for t,w in zip(ising.asdict()['terms'],ising.asdict()['weights']):
#    print(t,':',w)

# plot the mdl_ising
print('linear terms:')
for var in mdl_ising.objective_expr.iter_terms():
    print(var)
print()
print('quad. terms:')
for quad_term in mdl_ising.objective_expr.iter_quads():
    print(quad_term)
```

### Creates the QAOA and solve Ising with brute force

```python
q = QAOA()

qiskit_device = create_device(location='local', name='vectorized')

q.set_device(qiskit_device)

q.set_circuit_properties(p=3, param_type='standard', init_type='ramp', mixer_hamiltonian='x')

q.set_backend_properties(prepend_state=None, append_state=None)

q.set_classical_optimizer(method='powell', maxfev=1000, tol=0.01,
                          optimization_progress=True, cost_progress=True, parameter_log=True)
q.compile(ising)

q.solve_brute_force(verbose=True)
res=q.brute_force_results['configuration']

visualize_results(res,len(containers),len(positions))
```
