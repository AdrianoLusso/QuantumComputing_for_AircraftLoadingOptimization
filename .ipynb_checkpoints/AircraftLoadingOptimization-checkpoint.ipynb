# Creating QUBO of Aircraft Loading Optimization (ALO)

```python
import docplex
from docplex.mp.model import Model

from openqaoa.problems.converters import FromDocplex2IsingModel # function to translate docplex problems into Ising Hamiltonians
#from converter import FromDocplex2IsingModel
from openqaoa import QAOA, QUBO
from openqaoa.backends import create_device
from openqaoa.utilities import bitstring_energy

import numpy as np
```

T1 - medium --- $t_i=1$ --- $d_i=1$

T2 - small --- $t_i=1$ --- $d_i=0.5$

T2 - big --- $t_i=0.5$ --- $d_i=1$

* Para T1, toda la proporcion de su peso esta en una posicion ($t_i$), y tambien ocupa una posicion entera ($d_i$).
* Para T2, toda la proporcion de su peso esta en una posicion ($t_i$),pero ocupa la mitad de una posicion porque es chiquitito ($d_i$).
* Para T3, la mitad de la proporcion de su peso esta en una posicion y la otra mitad en otra ($t_i$),y ocupa una posicion entera ($d_i$).

IMPORTANTE, EL PESO MAXIMO DEBE SER 1, Y LOS CONTAINER MASS VALORES ENTRE 0 A 1, POR LO TANTO, HAY QUE NORMALIZAR LA MASAS Y PESO MAXIMOS REALES A METRICAS ENTRE 0 A 1.

```python
containers = list(range(3))
containers_type = [2,1,3]
containers_mass = [0.22 , 0.23 , 0.21]

positions = list(range(3))

containers_proportion_of_mass_per_position = [0.5 if type == 3 else 1 for type in containers_type]
containers_occupacy_per_position = [0.5 if type == 2 else 1 for type in containers_type]

max_weight = 1
allBinaryVariables = list(range(len(containers) * len(positions)))
```

```python
mdl = Model("ALO")
x = mdl.binary_var_list(len(allBinaryVariables), name="x")

# OBJECTIVE FUNCTION + CONTIGUITY OF BIG CONTAINERS
mdl.minimize(
    # OBJECTIVE FUNCTION
    -mdl.sum(
        t_i * m_i * x[len(positions) * i + j] 
        for j in positions
        for t_i,m_i,i in zip(containers_proportion_of_mass_per_position,containers_mass,containers)
    )
    
)

'''
+2* # CONTIGUITY OF BIG CONTAINERS
    mdl.sum(
        0.5 * mdl.sum(
                x[len(positions) * i + j]
                for j in positions
            )
        -
        mdl.sum(
            x[len(positions) * i + j] * x[len(positions) * i + (j+1)]
            for j in positions[:-1]
        )
        for i in containers
        if containers_type[i]==3
    )
'''

# NO OVERLAPS
for j in positions:
    mdl.add_constraint(mdl.sum(
        d_i * x[len(positions) * i + j]
        for d_i,i in zip(containers_occupacy_per_position,containers) 
        ) == 0.5
        ,ctname="no overlaps"
    )

# NO DUPLICATES
for i in containers:
    t_i = containers_proportion_of_mass_per_position[i]
    mdl.add_constraint(t_i * mdl.sum(
        x[len(positions) * i + j]
        for j in positions 
        ) == 0.5
        ,ctname="no duplicates"
    )
    
# CONTIGUITY OF BIG CONTAINERS
for i in containers:
    if containers_type[i]!=3:
        continue
    mdl.add_constraint(
        0.5 * mdl.sum(
                x[len(positions) * i + j]
                for j in positions
            )
        ==
        mdl.sum(
            x[len(positions) * i + j] * x[len(positions) * i + (j+1)]
            for j in positions[:-1]
        ) 
        ,ctname="contiguity"
    )

# MAXIMUM CAPACITY
mdl.add_constraint(
    mdl.sum(
        t_i * m_i * x[len(positions) * i + j] 
        for j in positions
        for t_i,m_i,i in zip(containers_proportion_of_mass_per_position,containers_mass,containers)
    ) <= max_weight
    ,ctname="maximum capacity"
)
#mdl.add_constraint(mdl.sum(var * weights[i] for i, var in enumerate(x)) <= max_weight)
```
